### Import library dan dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv("DatasetStudiKasus.csv")

print(dataset)
```

### 1.a) Berapa rata-rata mahasiswa datang pada minggu ini?

```python
rata_rata_datang = dataset['Datang'].mean()
print("Rata-rata mahasiswa datang pada minggu ini:", rata_rata_datang)
```

### 1.b) Kapan biaya tertinggi terjadi?

```python
dataset['TotalBiaya'] = dataset['Biaya'] * dataset['Datang']
biaya_tertinggi = dataset.loc[dataset['TotalBiaya'].idxmax()]
print("Biaya tertinggi terjadi pada:\n", biaya_tertinggi)
```

### 1.c) Hari apa biaya lebih dari 1100000

```python
dataset['TotalBiaya'] = dataset['Biaya'] * dataset['Datang']
biaya_lebih_dari_110000 = dataset[dataset['TotalBiaya'] > 110000]['Hari'].unique()
print("Biaya lebih dari 110000:", biaya_lebih_dari_110000)
```

### 1.d) Siapa yang paling banyak datang ke kampus?

```python
paling_banyak_datang_ke_kampus = dataset.groupby('Mahasiswa')['Datang'].sum().idxmax()
print("Paling banyak datang ke kampus:", paling_banyak_datang_ke_kampus)
```

### 1.e) Siapa yang datang pada hari minggu?

```python
datang_pada_hari_minggu = dataset[dataset['Hari'] == 'Minggu']['Mahasiswa'].unique()
print("Datang pada hari minggu:", datang_pada_hari_minggu)
```

### 1.f) Berapa biaya tertinggi dan terendah?

```python
dataset['TotalBiaya'] = dataset['Biaya'] * dataset['Datang']
biaya_tertinggi = dataset['TotalBiaya'].max()
biaya_terendah = dataset['TotalBiaya'].min()
print("Biaya tertinggi:", biaya_tertinggi, " dan terendah:", biaya_terendah)
```

### 1.g) Berapa frekuensi datang tertinggi dan terendah?

```python
datang_tertinggi = dataset['Datang'].max()
datang_terendah = dataset['Datang'].min()
print("Frekuensi datang tertinggi:", datang_tertinggi, " dan terendah:", datang_terendah)
```

```python
from plotnine import ggplot, aes, geom_bar, labs, theme_minimal, theme, element_text
import pandas as pd

fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]

info_mahasiswa = pd.DataFrame({
    'fakultas': fakultas,
    'jumlah_mahasiswa': jumlah_mahasiswa,
    'akreditasi': akreditasi
})

gambar = (ggplot(info_mahasiswa, aes(x='fakultas', y='jumlah_mahasiswa', fill='fakultas'))
          + geom_bar(stat='identity', width=0.8)
          + theme_minimal()
          + labs(title='Jumlah Mahasiswa per Fakultas', x='Fakultas', y='Jumlah Mahasiswa')
          + theme(axis_text_x=element_text(angle=45, hjust=1))
         )

print(gambar)
```
